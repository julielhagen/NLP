```python
import pandas as pd
```

```python
gold_eng = pd.read_csv("ENg_ANN_golddata_fix.csv", sep=";")
columns_to_drop = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']
gold_eng.drop(columns=columns_to_drop, inplace=True)
gold_eng.drop(gold_eng.tail(1).index, inplace=True)
gold_eng
```

```python
gold_words = gold_eng["Sentences"]
gold_labels = gold_eng["Labels"]
```

```python
gold_labels_lst = gold_labels.values.tolist()
gold_words_lst = gold_words.values.tolist()
```

```python
gold_words_lst
```

```python
label_indx_lists = []

# Iterate over each list in the list of lists
for inner_list in gold_words_lst:
    indexed_inner_list = []
    # Enumerate over each element in the inner list
    for idx, value in enumerate(inner_list):
        indexed_inner_list.append((idx+1))  # Append a tuple of index and value
    label_indx_lists.append(indexed_inner_list)  # Append the indexed inner list to the result

print(len(label_indx_lists[100]))
```

```python
# Count the total number of numbers
total_numbers = sum(len(sublist) for sublist in label_indx_lists)

print("Total number of numbers:", total_numbers)
```

```python
indexed_word_lists = []

# Iterate over each list in the list of lists
for inner_list in gold_words_lst:
    indexed_inner_list = []
    # Enumerate over each element in the inner list
    for idx, word in enumerate(inner_list):
        indexed_inner_list.append((idx+1))  # Append a tuple of index and word
    indexed_word_lists.append(indexed_inner_list)  # Append the indexed inner list to the result
len(indexed_word_lists[100])
```

```python
# Count the total number of numbers
total_numbers_words = sum(len(sublist) for sublist in indexed_word_lists)

print("Total number of numbers:", total_numbers_words)
```
